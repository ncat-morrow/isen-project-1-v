## Histogram Plotting
Plot a histogram for each of the three features of the dataset. Be sure to include proper titles and axis labels on each of your plots.

```python
df[selected_columns].hist()
plt.show()
```

```python
plt.hist(df['Outside Air Temperature'], bins=20, color='skyblue')
plt.title('Histogram of Outside Air Temperature')
plt.xlabel('Temperature (°F)')
plt.ylabel('Frequency')
```

```python
plt.hist(df['Heating Water Flow Rate'], bins=20, color='lightgreen')
plt.title('Histogram of Heating Water Flow Rate')
plt.xlabel('Flow Rate')
plt.ylabel('Frequency')
```

```python
plt.hist(df['TotalEnergy'], bins=20, color='lightcoral')
plt.title('Histogram of Total Energy')
plt.xlabel('Energy (Units)')
plt.ylabel('Frequency')
```
